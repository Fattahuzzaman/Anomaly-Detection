```python
dft=df[['Date','Price']]
dft.head()
```

```python
dft['Date'] = pd.to_datetime(df['Date'])
dft.head()
```

```python
dft.index_col='Date'
dft.head()
```

```python
sns.lineplot(x=dft['Date'], y=dft['Price'])
print("Start date is: ", dft['Date'].min())
print("End date is: ", dft['Date'].max())
```

```python
train, test = dft.loc[dft['Date'] <= '2019-06-30'], dft.loc[dft['Date'] > '2019-06-30']
```

```python
scaler=StandardScaler()
scaler = scaler.fit(train[['Price']])

train['Price'] = scaler.transform(train[['Price']])
test['Price'] = scaler.transform(test[['Price']])
```

```python
seq_size=30
def to_sequences(x, y, seq_size=1):
    x_values = []
    y_values = []

    for i in range(len(x)-seq_size):
        #print(i)
        x_values.append(x.iloc[i:(i+seq_size)].values)
        y_values.append(y.iloc[i+seq_size])
        
    return np.array(x_values), np.array(y_values)

trainX, trainY = to_sequences(train[['Price']], train['Price'], seq_size)
testX, testY = to_sequences(test[['Price']], test['Price'], seq_size)
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(trainX.shape[1], trainX.shape[2])))
model.add(Dropout(rate=0.2))

model.add(RepeatVector(trainX.shape[1]))

model.add(LSTM(128, return_sequences=True))
model.add(Dropout(rate=0.2))
model.add(TimeDistributed(Dense(trainX.shape[2])))
model.compile(optimizer='adam', loss='mae')
model.summary()
```

```python
history = model.fit(trainX, trainY, epochs=30, batch_size=32, validation_split=0.1, verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
```

```python
trainPredict = model.predict(trainX)
trainMAE = np.mean(np.abs(trainPredict - trainX), axis=1)
plt.hist(trainMAE, bins=30)
```

```python
max_trainMAE = 0.8  #or Define 90% value of max as threshold.

testPredict = model.predict(testX)
testMAE = np.mean(np.abs(testPredict - testX), axis=1)
plt.hist(testMAE, bins=30)
```

```python
anomaly_df = pd.DataFrame(test[seq_size:])
anomaly_df['testMAE'] = testMAE
anomaly_df['max_trainMAE'] = max_trainMAE
anomaly_df['anomaly'] = anomaly_df['testMAE'] > anomaly_df['max_trainMAE']
anomaly_df['Price'] = test[seq_size:]['Price']
```

```python
sns.lineplot(x=anomaly_df['Date'], y=anomaly_df['testMAE'])
sns.lineplot(x=anomaly_df['Date'], y=anomaly_df['max_trainMAE'])
```

```python
anomalies = anomaly_df.loc[anomaly_df['anomaly'] == True]

#Plot anomalies
sns.lineplot(x=anomaly_df['Date'], y=scaler.inverse_transform(anomaly_df['Price']))
sns.scatterplot(x=anomalies['Date'], y=scaler.inverse_transform(anomalies['Price']), color='r')
```
